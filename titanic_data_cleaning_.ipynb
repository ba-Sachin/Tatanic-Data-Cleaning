```python
import pandas as pd
from sklearn.impute import SimpleImputer # Import SimpleImputer from sklearn.impute
from sklearn.preprocessing import StandardScaler, LabelEncoder # standardscaler should be StandardScaler
```

```python
df= pd.read_excel('/Titanic Data Cleaning.xlsx')
```

```python
df
```

```python
(df[['Age','Cabin','Embarked']]).isnull().sum()
```

```python

import numpy as np
df[['Age','Cabin','Embarked']]=df[['Age','Cabin','Embarked']].astype(object).where(pd.notnull(df[['Age','Cabin','Embarked']]),np.nan)
```

```python
df
```

```python
age_imputer = SimpleImputer(strategy ='mean')
cabin_imputer = SimpleImputer(strategy = 'most_frequent')
embarked_imputer = SimpleImputer(strategy = 'most_frequent')
```

```python
# apply .imputer with .ravel() to flatten the output
df['Age'] = age_imputer.fit_transform(df[['Age']])
df['Cabin'] = cabin_imputer.fit_transform(df[['Cabin']]).ravel()
df['Embarked'] = embarked_imputer.fit_transform(df[['Embarked']]).ravel()
```

```python
# Encode Sex: male -> 0, female -> 1
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
```

```python
df['Fmailysize'] = df['SibSp']+df['Parch']+1
```

```python
df
```

```python
# Define function to categorize Age
def categorize_age(age):
    if age < 18:
        return 'Child'
    elif age < 60:
        return 'Adult'
    else:
        return 'Senior'

# Apply it to create a new column
df['AgeGroup'] = df['Age'].apply(categorize_age)
```
